```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
baseline_df = pd.read_csv('baseline.txt', delimiter=' ', header=None)
```

```python
baseline_df[0][0]
baseline_df[0][1]
```

```python
df_random_1 = pd.read_csv('randomDetectorsP08I10.txt', delimiter=' ', header=None)
df_random_2 = pd.read_csv('randomDetectorsP08I20.txt', delimiter=' ', header=None)
df_social_1 = pd.read_csv('socialDetectorsP08I10.txt', delimiter=' ', header=None)
df_social_2 = pd.read_csv('socialDetectorsP08I20.txt', delimiter=' ', header=None)
```

```python
def normalizeReduction(row, baseline):
    y = row[1]
    normalized = (baseline - y)/baseline
    #print normalized
    return normalized
```

```python
df_random_1[5] = df_random_1.apply(normalizeReduction, axis=1,  baseline=baseline_df[0][0])
df_random_2[5] = df_random_2.apply(normalizeReduction, axis=1,  baseline=baseline_df[0][1])
df_social_1[5] = df_social_1.apply(normalizeReduction, axis=1,  baseline=baseline_df[0][0])
df_social_2[5] = df_social_2.apply(normalizeReduction, axis=1,  baseline=baseline_df[0][1])
```

```python
print df_social_1[5]
print df_random_1[5]
```

```python
plt.figure(figsize=(12,12))
plt.subplot(221)
plt.title('Reduction in infected nodes:\nnumber of intial infected nodes = 1%\n infect probability=0.8')
plt.xlabel("Number of detectors")
plt.ylabel("reduction in infected nodes")
x = df_random_2[0].tolist()
y = df_random_2[5].tolist()
plt.plot(x, y, '-bo')
y_s = df_social_2[5].tolist()
plt.plot(x, y_s, '-g*')
plt.legend(['Randomly selected Detector Nodes', 'Nodes with high Social Score as Detectors'], loc=4)
plt.subplot(222)
plt.title('Reduction in infected nodes:\nnumber of intial infected nodes = 0.5%\n infect probability=0.8')
plt.xlabel("Number of detectors")
plt.ylabel("reduction in infected nodes")
x = df_random_1[0].tolist()
y = df_random_1[5].tolist()
plt.plot(x, y, '-bo')
y_s = df_social_1[5].tolist()
plt.plot(x, y_s, '-g*')
plt.legend(['Randomly selected Detector Nodes', 'Nodes with high Social Score as Detectors'],loc=4)
plt.savefig('reductionInInfected.png')
plt.show()
```
